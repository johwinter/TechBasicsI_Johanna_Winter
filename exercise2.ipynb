### Variables

In math, a variable is a letter that represents an unknown number. Ex: `1 + x  = 4.` In programming, variables are more like holders that can store any information that will be needed later in the program's execution. The value stored in the variable changes each time the variable is assigned a new value.

In Python, variables are more flexible than programming languages such as C. It is possible to assign different data types to the same variable and it's type changes automatically. You can use the function `type()` to find out the data type of any expression.


```python
x = 1
print("The current value of X is: ", x, ". Data type is: ", type(x))
x = "amazing"
print("The current value of X is: ", x, ". Data type is: ", type(x))
x = True
print("The current value of X is: ", x, ". Data type is: ", type(x))
x= [1,2,3]
print("The current value of X is: ", x, ". Data type is: ", type(x))
```

You can also convert data types based on your need

```python
result = 13 * 11
print(result, type(result))

result = str(result) # here we are using str() to convert result from int to string
print(result, type(result))
```

Variables are also useful to take user inputs and save them for later.

```python
something = input("Type something: ")
print("You typed:", something)
print(type(something))
```

Notice inputs are by default strings. If we need users to input some numbers, we will need to manually convert them into int in the code.

```python
number1 = input("Enter the first number: ")
number2 = input("Enter the second number: ")
print(number1 + number2)
print(int(number1) + int(number2))
```

We learnt that `=` is an assignment operator and  `==` really means equal. There are more special assignment operators that combines assignment with algorithmic operators. What should the following code print out?


```python
x = 1
x += 3
print(x)
x -= 1
print(x)
```

### More about Print statement

We have been defining strings with either double quotes or single quotes. What if you want to print both double quotes and single quotes in one string? In such cases you can *escape* special characters by putting a backslash `\` before. For example:

```python
print("The teacher said. \"I think that's a cool idea!\"")
```

See how the escaped characters got rendered in a different color in your code editor.

They are called escape characters because they are escaped from their usual meaning. There are some other useful escape characters that are commonly used:

- `\n` for new line in your string
- `\t` for adding a tab in your string
- `\\` you will need to escape your backslash if you want to print one

How many backslash will you get in the end?

```python
print("\\\\\\\\\n\\\\\\\\")
```

Some algorithmic Operations can also be applied to strings

```python
print("!" * 3)
print("a" + "b")
print("10" + "10")
```

Strings have length, and you can use `len()` to get  the length of your string.

```python
print(len("How many characters are there in this sentence?)"))
print(len("\n"))
```

You can also transform a string to make sure you only have uppercase or lowercase in your string.

```python
print('uppercase'.upper())
print('Lowercase'.lower())
```

Another very useful trick that can be combined with print command is the f-string, formatted string. A formatted string looks like this:

```python
program = "Digital Media"
semester = 2
print(f"I study {program} in semester {semester}.")
```

By using the f-string we can print a string with multiple variables without worrying about their data type.

### Conditional Statements

Conditional statements are there for us to control the flow of execution by evaluating whether a condition is true or false. Everything indented the same amount after the colon (:) will be executed whenever the condition is true.

```python
season = "winter"
mood = "Feeling okay."

if season == "summer":
    print("Time to enjoy the sun.")
    mood = "Feeling happy."

print(mood)
```

 After the colon we always need to have a line break and indent the next line of code. If you type an if statement in PyCharm, once you hit enter, the indent is automatically added for you. Just to show the wrong example, the same syntax without indent won't work in Python:

```python
if season := "summer":
    print("Time to enjoy the sun.")
```

You can chain multiple conditions up like this:

```python
season = "summer"
mood = "good"

if season == "summer" or mood == "good":
    print("Let's go out!")

if season == "winter" and mood == "bad":
    print("Let's stay home.")
```

An if statement can also be combined with an else statement, which handles all the other cases when the if statement is not fulfilled.

```python
temperature = 5

if temperature > 0:
    print("no ice")
else:
    print("frozen")
```

One can also nest multiple conditional statements. Correct indent is very important here for the parser to know which statements go together.

```python
temperature = -1

if temperature > 0:
    if temperature >= 30:
        print("too hot")
    elif temperature < 10:
        print("too cold")
    else:
        print("comfortable")
else:
    print("frozen!")
```

Conditional statements are very handy when handling user input.

```python
number = int(input("Please type in a number: "))

if number < 0:
    print("The number is negative.")
elif number == 0:
    print("The number is zero.")
else:
    print("The number is positive.")
```

### Debugging
Very often when we test the code we wrote, we encounter error messages:

```python
x = 10
y = 2
result = x / y

print(f"{x} divided by {y} is {result}")
```

It is important to stay calm and read what is being written in the error message. Usually it states the line of code as well as the reason that is causing the error.

But there are other cases when the program runs without errors, but they don't really do what we intend them to do :/. Now what we need is <i>debugging</i>: the process discovering the causes of bugs. People who code often spend more time debugging than writing fresh code.

A simple yet effective way of debugging is adding print statements to your code. With print statements, we can check the current state of variables and see if they are actually behaving as what we have in mind.



```python
# let's say we have a little game
growingRate = 1
season = "summer"

if season == "Summer":
    growingRate * 2

print("Growing rate is:", growingRate)
```

Seems like our growing rate is not doubled as we expected. What debugging print statements can we add here?

```python
# let's say we have a little game
growingRate = 1
season = "summer"

print("Condition:", season == "Summer")
if season == "Summer":
    growingRate * 2

print("Growing rate is:", growingRate)
```

By checking the value of the condition, we realize that the condition is never met. Now it is also easier to notice that "summer" and "Summer" are not the same, so we need to change one of the strings to make this work. After fixing the condition, the code is still behaving the same. Now it's time to add another print statement inside the if statement to check what's going on with `growingRate`.

```python
# let's say we have a little game
growingRate = 1
season = "summer"

# print("Condition:", season == "Summer")
if season == "summer":
    growingRate * 2
    print(growingRate)

print("Growing rate is:", growingRate)
```

Checking the output, we can say that the program can finally go inside the if statement. The value of growingRate has not been changed here because we never assigned the doubled value to the variable.

```python
# let's say we have a little game
growingRate = 1
season = "summer"

# print("Condition:", season == "Summer")
if season == "summer":
    growingRate = growingRate * 2
    # print(growingRate)

print("Growing rate is:", growingRate)
```

After fixing everything, remember to remove the print statements for debugging purpose. Simple bugs like this one might also be identified easily just by reading carefully. This is a simple demo to show when and how printing statements can be used for debugging.
